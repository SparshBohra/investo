1. Simple Moving Average Crossover for Hindalco

```python
# function to calculate sma
def SMA(data, period=30, column='close'):
    return data[column].rolling(window=period).mean()
```

```python
# saving the 20 and 50 day sma
data['SMA20'] = SMA(data, 20)
data['SMA50'] = SMA(data, 50)
```

```python
# buy and sell signals
data['Signal'] = np.where(data['SMA20'] > data['SMA50'], 1, 0)
data['Position'] = data['Signal'].diff()

data['Buy'] = np.where(data['Position'] == 1, data['close'], np.NAN)
data['Sell'] = np.where(data['Position'] == -1, data['close'], np.NAN)
```

```python
# Visualise closing price with SMAs and Buy & Sell signals
plt.figure(figsize=(16, 8))
plt.title('Closing Prices with Buy/Sell Signals', fontsize=18)

plt.plot(data['close'], alpha=0.5, label='Close')
plt.plot(data['SMA20'], alpha=0.5, label='SMA20')
plt.plot(data['SMA50'], alpha=0.5, label='SMA50')

plt.scatter(data.index, data['Buy'], alpha=1, label='Buy Signal', marker='^', color='green')
plt.scatter(data.index, data['Sell'], alpha=1, label='Sell Signal', marker='v', color='red')

plt.xlabel('Date', fontsize=14)
plt.ylabel('Closing Price', fontsize=14)
plt.show()
```

2. Moving Average Crossover (with backtesting) for Yahoo Finance 

```python
# define variables
FAST_MA = 50
SLOW_MA = 200
STARTING_BALANCE = 10000
```

```python
# define time period
START = datetime.datetime(2014, 1, 1)
END = datetime.datetime(2019, 1, 1)
YEARS = (END - START).days / 365.25
```

```python
#load data into a pandas dataframe
yf.pdr_override()
price = pdr.get_data_yahoo('^GSPC', START, END)
price.head()
```

```python
# dropping redundant features
data2 = price[['Open', 'Close']]
data2.columns = ['open', 'close']
data2
```

```python
# plot chart
plt.plot(data2.close)
plt.show()
```

```python
# calculate daily return
data2['Return'] = data2.close / data2.open
data2['Bench_Bal'] = STARTING_BALANCE * data2.Return.cumprod()
data2.head()
```

```python
# calculate metrics
bench_return = round(((data2.Bench_Bal[-1] / data2.Bench_Bal[0]) - 1) * 100, 2)
bench_cagr = round((((data2.Bench_Bal[-1] / data2.Bench_Bal[0]) ** (1/YEARS))-1) * 100, 2)

print(bench_return)
print(bench_cagr)
```

```python
# calculate drawdown
data2['Bench_Peak'] = data2.Bench_Bal.cummax()
data2['Bench_DD'] = data2.Bench_Bal - data2.Bench_Peak

bench_dd = round((((data2.Bench_DD / data2.Bench_Peak).min()) * 100), 2)
bench_dd
```

```python
# calculate moving averages
data2['Fast_MA'] = data2.close.rolling(window = FAST_MA).mean()
data2['Slow_MA'] = data2.close.rolling(window = SLOW_MA).mean()
data2.tail()
```

```python
data2['Signal'] = np.where(data2['Fast_MA'] > data2['Slow_MA'], 1, 0)
data2['Position'] = data2['Signal'].diff()

data2['Buy'] = np.where(data2['Position'] == 1, data2['close'], np.NAN)
data2['Sell'] = np.where(data2['Position'] == -1, data2['close'], np.NAN)

# Visualise closing price with SMAs and Buy & Sell signals
plt.figure(figsize=(16, 8))
plt.title('Closing Prices with Buy/Sell Signals', fontsize=18)

plt.plot(data2.close, alpha=0.5, label='Close')
plt.plot(data2.Fast_MA, alpha=0.5, label='Fast_MA')
plt.plot(data2.Slow_MA, alpha=0.5, label='Slow_MA')

plt.scatter(data2.index, data2['Buy'], alpha=1, label='Buy Signal', marker='^', color='green')
plt.scatter(data2.index, data2['Sell'], alpha=1, label='Sell Signal', marker='v', color='red')

plt.xlabel('Date', fontsize=14)
plt.ylabel('Closing Price', fontsize=14)
plt.show()
```

```python
# define entries
data2['Long'] = data2.Fast_MA > data2.Slow_MA
data2.tail()
```

```python
# calculate system balance
data2['Sys_Ret'] = np.where(data2.Long.shift(1) == True, data2.Return, 1)
data2.tail()
```

```python
# calculate system balance
data2['Sys_Bal'] = STARTING_BALANCE * data2.Sys_Ret.cumprod()
data2.tail()
```

```python
plt.plot(data2.Bench_Bal)
plt.plot(data2.Sys_Bal)
plt.show()
```

RESULTS

```python
# calculate metrics
sys_return = round(((data2.Sys_Bal[-1] / data2.Sys_Bal[0]) - 1) * 100, 2)
sys_cagr = round((((data2.Sys_Bal[-1] / data2.Sys_Bal[0]) ** (1/YEARS))-1) * 100, 2)

print(sys_return)
print(sys_cagr)
```

```python
# calculate drawdown
data2['Sys_Peak'] = data2.Sys_Bal.cummax()
data2['Sys_DD'] = data2.Sys_Bal - data2.Sys_Peak
sys_dd = round((((data2.Sys_DD / data2.Sys_Peak).min()) * 100), 2)

print(sys_dd)
```

```python
print(f'Benchmark Total return: {bench_return}%')
print(f'Benchmark CAGR: {bench_cagr}')
print(f'Benchmark DD: {bench_dd}%')
print('')
print(f'System Total return: {sys_return}%')
print(f'System CAGR: {sys_cagr}')
print(f'System DD: {sys_dd}%')
```
